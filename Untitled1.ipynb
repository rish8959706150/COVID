```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np

```

```python
df= pd.read_csv('Desktop/CSV fies/covid_19_india.csv',parse_dates=['Date'],dayfirst = True)
```

```python
df.head()
```

```python
today = df [df.Date == '2021-8-11']
```

```python
today.head()
```

```python
today.tail()
```

```python
max_confirmed = today.sort_values (by = "Confirmed",ascending=False)
```

```python
top_states = max_confirmed [0:5]
```

```python
top_states
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="State/UnionTerritory",y="Confirmed",data=top_states ,hue="State/UnionTerritory")
plt.show()
```

```python
max_deaths=today.sort_values (by = "Deaths",ascending=False)
```

```python
top_deaths=max_deaths[0:5]
top_deaths
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="State/UnionTerritory",y="Deaths",data=top_deaths ,hue="State/UnionTerritory")
plt.show()
```

```python
max_cured=today.sort_values (by = "Cured",ascending=False)
```

```python
top_cured=max_cured[0:5]
top_cured
```

```python

sns.barplot(x="State/UnionTerritory",y="Cured",data=top_cured ,hue="State/UnionTerritory")
plt.show()
```
